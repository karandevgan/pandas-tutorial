```python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
df = pd.read_csv('PS65_1337058552336.csv')
```

```python
df.columns = ['SNo', 'State', 'MaleFull', 'FemaleFull', 'TotalFull', 'MaleContract', 'FemaleContract', 'TotalContract']
```

```python
df.index = df['SNo']
del df['SNo']
```

```python
df.head()
```

```python
df['MaleFullPercent'] = (df['MaleFull'] / df['TotalFull']) * 100
```

```python
df
```

```python
MaleStateGroup = pd.DataFrame(df['MaleFullPercent'].groupby(df['State'], as_index=True).sum())
```

```python
MaleStateGroup['MaleFullPercent'].plot(figsize=(15,10), kind='bar')
```
